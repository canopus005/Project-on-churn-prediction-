```python
df=df.drop(['CustomerID','Name'],axis=1)
```

```python
df.head()
```

```python
cat = []
con = []
for i in df:
    if(df[i].dtypes == "object"):
        cat.append(i)
    else:
        con.append(i)
```

```python
len(cat)
```

```python
len(con)
```

```python
X1=pd.get_dummies(df[cat])
X1
```

```python
from sklearn.preprocessing import MinMaxScaler
Min_scaler=MinMaxScaler()
array=Min_scaler.fit_transform(df[con])
X2=pd.DataFrame(array,columns=con)
X2
```

```python
Xnew=X2.join(X1)
Xnew
```

```python
from sklearn.model_selection import train_test_split,RandomizedSearchCV,GridSearchCV,cross_val_score
from sklearn.metrics import accuracy_score
X=Xnew.drop(['Churn'],axis=1)
X
```

```python
Y=df[['Churn']]
Y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(X,Y,test_size=0.25,random_state=31)
xtrain
```

```python
xtest
```

```python
from sklearn.linear_model import LogisticRegression
Log_model=LogisticRegression()
Log_model.fit(xtrain,ytrain)
```

```python
#training accuracy 
yprdtrain=Log_model.predict(xtrain)

Acc=accuracy_score(ytrain,yprdtrain)
print("Accuracy :",Acc)

CV_Score=cross_val_score(Log_model,X,Y,cv=5)
print('CV_Score :',np.mean(CV_Score))

# testing accuracy
yprdtest=Log_model.predict(xtest)

Acc=accuracy_score(ytest,yprdtest)
print("Accuracy :",Acc)

CV_Score=cross_val_score(Log_model,X,Y,cv=5)
print('CV_Score :',np.mean(CV_Score))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model=KNeighborsClassifier()
knn_model.fit(xtrain,ytrain)
```

```python
#TRAINING
ypredtrain=knn_model.predict(xtrain)
accuracy=accuracy_score(ytrain,ypredtrain)
print('Accuracy :',accuracy*100)
score=cross_val_score(knn_model,X,Y,cv=5)
print('cross validation value',np.mean(score))

# TESTING
ypredtest=knn_model.predict(xtest)
accuracy=accuracy_score(ytest,ypredtest)
print('Accuracy :',accuracy*100)
score=cross_val_score(knn_model,X,Y,cv=5)
print('cross validation value',np.mean(score))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(random_state=15)
rf_model.fit(xtrain,ytrain)
```

```python
#TRAINING
ypredtrain=rf_model.predict(xtrain)
accuracy=accuracy_score(ytrain,ypredtrain)
print('Accuracy :',accuracy*100)
score=cross_val_score(rf_model,X,Y,cv=5)
print('cross validation value',np.mean(score))

# TESTING
ypredtest=rf_model.predict(xtest)
accuracy=accuracy_score(ytest,ypredtest)
print('Accuracy :',accuracy*100)
score=cross_val_score(rf_model,X,Y,cv=5)
print('cross validation value',np.mean(score))
```
